```python
# Criando uma lista em python obs: a lista adminite strings com comandos 
list = [0, 'eu me amo', 2.5]
print(list)

# Nesta parte fiz a adição entre duas strings, sendo que list [1] é o segundo indice do vetor list 
print('variavel:'+list[1])

# Adição do primeiro indice da variavel list (list[0]) - Lembrar que a indexação no pythin inicia em 0
print(list[0]+1)
```

```python
# Criando uma matriz numpy
import numpy as np

# Criando uma Matriz de zeros utilizando np.zeros 
a = np.zeros((5,5))
print(a)
print('indexação numpy')
print(a[0,0])
```

```python
# Criando Numeros Aleatórios e mostrando a dimensão da matriz 
b = np.random.rand(5,5)
print(b)
print('DImensao da matriz')
print(b.shape)
print(type(b))
```

```python
# Criando matriz de valores aleatorios entre uma faixa estabelecida 

vmin = 1 #valor minimo 
vmax = 100 #valor maximo 

# Matriz
c = vmin+(vmax-vmin)*b
print(c)
```

```python
#plotar matriz 

import matplotlib.pyplot as plt 

plt.plot(c.transpose())
```

```python
# plotando apenas a primeira linha 
x = np.array(range(0,np.size(c[0,:])))
print(x)
plt.plot(c[0,:])

```

```python
# criando um grafico de barras e exportando

fig, ax = plt.subplots(3,1)
ax[0].bar(x,c[:,0])
ax[1].bar(x,c[:,1])
ax[2].bar(x,c[:,2])

# exportar a figura
import os
rootFolder = os.getcwd()
print(rootFolder)
figFolder = rootFolder.replace('\\','/')+'/figures/'
print(figFolder)

fig.savefig(figFolder+'Figure01.png')
```
